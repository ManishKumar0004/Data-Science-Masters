```python
"""Question-1:Explain why we have to use the Exception class while creating a Custom Exception ?
              Note:- Here Exception class refers to the base class for all the exception ?
   Ans:- 
   
   In python , we can define custom exceptions by creating a new clas that is derived from the built-in
   Exception class. """
   
   # Here's the syntax to define custom exceptions :
   
""" 
   
   class CustomError(Exception):
      ...
      pass
      
   try :
      
      ...
      
   except :
   
      ...
      
"""

""" Here, CustomError is a user-defined error which inherits from the Exception class.

Note :- 

When we are developing a large Python program,it is a good practice to place all the user-defined exceptions that
our program riases in a seperate file.

Many standard modules define their exceptions separately as exceptions.py or error.py (generally but not always). """

# Example :-

# define Python user-defined exceptions

class InvalidAgeException(Exception):
    " Raised when the input value is less than 18 "
    pass

# you need to guess this number
number=18

try:
    
    input_num=int(input("Enter a number:"))
    if input_num < number:
        raise InvalidAgeException
    else :
        print("Eligible to vote")
except InvalidAgeException :
    print("Exception Occurred : Invalid Age")
```

```python
"""Question-3: What errors are defined in the ArithmeticError class ? Explain any two with an example ?
   Ans:- 
   
   ArithmeticError class is the base class for those built-in exceptions that are raised for various arithmetic
   errors such as :
   (1) OverflowError
   (2) ZeroDivisionError
   (3) FloatingPointError             """

# Example 1 :-

import logging 
logging.basicConfig(filename="error.log", level=logging.ERROR)
  
try :
    a=10/0
    print(a)
except ZeroDivisionError as e:
    logging.error("i am trying to handle a zerodivision error {}".format(e))
    print("The statement is raising an ZeroDivisionError")
    print(e)
```

```python
# Example 2 :-
import math 

try :
    a=math.exp(1000)
    print(a)
except OverflowError as e :
    logging.error("i am trying to handle a overflow error {}".format(e))
    print("The statement is raising an arithmetic exception")
    print(e)
else :
    print("Success")
```

```python
"""Question-4 : Why LookupError class is used ? Explain with an example KeyError and IndexError ?
   Ans :-
   
   The LookupError exception in python forms the base class for all exceptions that are raised when an index or a key
   is not found fro a sequence or dictionary respectively.
   
   We can use LookupError exception class to handle both IndexError and KeyError exception classes. """

# Example (KeyError) :- 

player_info = {"name":"Manish", "age":20, "role":"All rounder"}
user_input=input("What do you want to know about the player")

try :
    
    print(f'{user_input} is {player_info[user_input]}')
except LookupError as e :
    logging.error("i am trying to handle a lookuperror {}".format(e))
    print("The statement is raising an key exception")
    print(f'{e}, {e.__class__}')
    print(e)
```

```python
# Example (IndexError) :-

# lists
l=[1,2,3,4,5]

try :
    print(l[10])
except IndexError as e :
    logging.error("i am trying to handle a lookuperror {}".format(e.__class__))
    print("The statement is raising an Index error")
    print(f'{e}, {e.__class__}')
    print(e)
```
