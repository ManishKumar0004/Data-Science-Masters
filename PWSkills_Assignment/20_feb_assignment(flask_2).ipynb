### Question-1: Explain GET and POST methods ?
### Ans:-
<b>(1) GET method: </b>

<b>It is the most common method which can be used to send data in the unencrypted form to the server.</b><br>

<b>GET sends data in the URL as a query string which is visible in the browser's address bar.</b><br>

<b>GET request is better for the data which does not need to be secure(it means the data <br> which does not contain images or word documents).</b><br>

<b>(2) POST method:</b><br>

<b>It is used to send the form data to the server. The server does not cache the data transmitted using the post method.</b><br>

<b>POST means through body I am sending you data and GET means through URL I am sending you data.</b><br>

<b>POST method is more secured because data is not been seen while in GET data is been seen in URL.</b><br>

<b>Example of POST and GET Method:</b><br>

<b>In this example we have HTML5 and CSS3 files and methods have been use to link form were we perform some basic operation of maths and we have render data through python file.</b><br>

<b>app.py</b>

```python
from flask import Flask, render_template, request,jsonify

app = Flask(__name__)

@app.route("/", methods=["GET", "POST"])
def home_page():
    return render_template("index.html")

@app.route('/math', methods=['POST'])
def math_operation():
    if request.method == "POST":
        ops = request.form["operation"]
        num1 = int(request.form["num1"])
        num2 = int(request.form["num2"])
        if ops == "add":
           r = num1 + num2
           user_result = "The sum of {} and {} is {}".format(num1, num2, r)
        if ops == "subtract":
           r = num1 - num2
           user_result = "The subtraction of {} and {} is {}".format(num1, num2, r)
        if ops == "multiply":
           r = num1 * num2
           user_result = "The multiplication of {} and {} is {}".format(num1, num2, r)
        if ops == "divide":
           r = num1 / num2
           user_result = "The division of {} and {} is {}".format(num1, num2, r)
        return render_template("results.html", result=user_result)
```

### Question-3: Why is redirect() used in Flask?
### Ans:-
<b>Flask redirect() : </b><br>

<b>Flask redirect is defined as a function or utility in Flask which allows developers to redirect users to a specified URL and assign a specified status code. When this function is called, a response object is returned, and the redirection happens to the target location with the specified status code.</b><br>

<b>When a website is visited, server receives a request and after processing responds to the request with a three-digit code
known as the HTTP code.</b><br>

<b>In a programming world, this status code that is returned as a response is known as Errors.</b><br>

<b>Errors and redirect function goes hand in hand.</b><br>

<b>Syntax of Flask redirect() :-</b><br>

<b>flask.redirect(location, status code, response)</b><br>

<b>In the above function :-</b><br>

<b>location: location parameter is the URL where response should be redirected.</b><br>

<b>status code: status code sent to browser's header defaults to 302 which is for ‘found’.</b><br>

<b>response : response parameter is used to instantiate response.</b><br>

<b>The jinja template library is used by flask to render templates.</b><br>

<b>Templates help you reuse visuals.This helps you divide the business logic (Python code) and presentation logic (HTML).</b><br>

<b>In your application, you will use templates to render HTML which will display in the user’s browser.</b><br>

<b>Jinja looks and behaves mostly like Python.
