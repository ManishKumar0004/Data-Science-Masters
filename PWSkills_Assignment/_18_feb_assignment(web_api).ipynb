#### Question-2: Give advantages and disadvantages of using API?
#### Ans:-

<b> The advantages of using APIs are: </b> <br>


<b> 1) Personalized Content :-</b>
API enables businesses to get a clear understanding of the tastes and preferences of the customer. This allows businesses to make personalized recommendations and generate better customer leads and interactions.<br>

<b> 2) Interoperability :- </b> APIs enable different software system to communicate and interact seamlessly,
even if they are built on different technologies. <br>

<b> 3) Automation:- </b>
One no longer needs to make assumptions and interpretations manually. API does it all for you by managing all the workflow making business processes more productive and efficient. <br>

<b> 4) Fraud Prevention:- </b>
API has various levels of security depending upon the type of API that you use which ensures that only authorized personnel have access to the data generated using API.

<b> 5) Improved Competency:- </b>
The information that is transmitted from one system to another is quickly intercepted and interpreted by API. This level of efficiency ensures information automatically available to every platform which leads to quicker and improved decision making.

<b>6) Adaptability:- </b>
API can anticipate changes in terms of needs that arise in the due course of business. It used the technology at its disposal to analyze the information present within the system more closely and thus makes the service provisions more flexible.


<b> The disadvantages of using APIs are: </b> <br>

<b> 1) Cost:- </b> <br>
Providing an API is expensive in terms of development time, ongoing maintenance, providing API documentation on your website and providing support to users of your API. <br>

<b> 2) Security exposure:- </b> <br>
Adding an API adds another attack surface to your website.

#### Question-3: What is a Web API? Differentiate between API and Web API.
#### Ans:-
A Web API is a set of standards and protocols for accessing web-based software applications or web tools. It provides a way for different software systems to communicate with each other over the internet.
It helps you to create and develop HTTP based RESTful services.
The primary purpose of a Web API is to allow different software systems to access and retrieve data from a remote server.

<b> Difference between API and Web API : </b> <br> 


 Here's the difference between API and Web API:

| Aspect           | API                                    | Web API                                        |
|------------------|----------------------------------------|------------------------------------------------|
| Definition       |An API is a Set of rules, protocols and tools that allow two software application to communicate with each other | A Web API is a special type of API that is designed to remote communication over the internet using standard web protocols such as HTTP |
| Accessibility    | Can be local or remote communication   | Specifically designed for remote communication over the internet |
| Communication Protocol | Various protocols, including non-web ones and TCP/IP,IPC| Primarily uses HTTP for web-based communication |
| Usage Context    | Various contexts (desktop, mobile, embedded system) | Commonly used in web and mobile app development |
| Location         | Local and remote                        | Hosted on web servers, accessible via URLs      |
| Access Method    | Method calls, function invocations      | HTTP requests (GET, POST, PUT, DELETE)          |
| Transport Layer  | Can operate at different layers         | Operates at the application layer using HTTP   |



#### Question-4: Explain REST and SOAP Architecture. Mention shortcomings of SOAP?
#### Ans:-

<b> REST Architecture:- </b> <br>


REST stands for REpresentational State Transfer and API stands for Application Program Interface.<br>
REST is a software architectural style that defines the set of rules to be used for creating web services.<br>
Web services which follow the REST architectural style are known as RESTful web services.<br>
It allows requesting systems to access and manipulate web resources by using a uniform and predefined set of rules.
Interaction in REST based systems happen through Internetâ€™s Hypertext Transfer Protocol (HTTP).

<b> A Restful system consists of a: </b> <br>
client who requests for the resources.<br>
server who has the resources.


<b> Architectural Constraints/guidlines of RESTful API: </b> <br>

There are six architectural constraints which makes any web service are listed below: <br>
i. Uniform Interface <br>
ii. Stateless <br>
iii. Cacheable <br> 
iv. Client-Server  <br>
v. Layered System <br>
vi. Code on Demand <br>

<b>SOAP Architecture:- </b> <br>

SOAP stands for Simple Object Access Protocol. <br>
SOAP is a protocol for sending and receiving messages(data) between applications over the internet. <br>
It uses the XML format of data to transfer messages over the HTTP protocol. <br> 
In Web services, SOAP allows the user request to interact with other programming languages. <br>
In this way, it provides a way to communicate between applications running on different platforms (Operating system), with programming languages and technologies used in web service.

<b> SOAP Building Block:- </b> <br>

The SOAP building block describes what XML data is sent to the web service and client application. The following diagram       represents the SOAP building block. <br>

<b> SOAP Envelope:</b> Envelope is used to define the start and end of the SOAP message. It contains the details of the SOAP message. It is an important element of the XML documents. <br>

<b>SOAP Header:</b> It is an optional element in which the header contains the credentials information such as authorization, authentication, etc. is used during the processing of a SOAP message. <br>

<b>SOAP Body: </b>It is an important element of the SOAP message that contains request and response information in the XML format. It defines the actual content of the message to be sent between the client and the webserver. <br>

<b>SOAP Fault: </b>The SOAP Fault element is an optional element used to display an error message encountered during the transmission of a SOAP message. It holds the status of SOAP messages and errors. <br>

<b> Shortcomings of SOAP:- </b> <br>

Here are four shortcomings of SOAP :

<b> 1. Complexity:</b> SOAP messages can be complex due to their XML structure, making development and debugging more challenging. <br>

<b> 2. Performance Overhead:</b> The XML format used in SOAP can lead to larger message sizes, increasing network overhead and slowing down communication. <br>

<b> 3. Limited Browser Support:</b> SOAP is not well-suited for direct use in web browsers, making it less ideal for client-side web applications.<br>

<b> 4. Less Human-Readability: </b> SOAP's XML structure is not as human-readable as other formats like JSON, making troubleshooting and understanding messages harder for developers.


#### question-5:Differentiate between REST and SOAP.
#### Ans:-

<b> Difference between REST and SOAP : </b>
| Aspect           | REST                                          | SOAP                                         
|------------------|-----------------------------------------------|------------------------------------------------|
| Definition | REST is an architectural style that uses HTTP protocol for communication     | SOAP is a protocol that defines a set of rules for structuring messages and uses various protocols for communication such as HTTP, SMTP etc.|
| State| REST is designed to be stateless, meaning each request from the client contains all the necessary information. | SOAP can have stateful or stateless communication, depending on how it's implemented.|
| Message Format | Generally transports data in JSON. It is based on URI.Because REST follows stateless model,REST does not enforces message format as XML or JSON etc.| Transports data in standard XML format.|
| Supported Method | It works with GET, POST, PUT, DELETE.| Because it is XML based and relies on SOAP, it works with WSDL.|
| Network Platforms | Works over HTTP and HTTPS.                    | Works over HTTP, HTTPS, SMTP, XMPP etc.|
| Scalability    | Scales well due to statelessness             | Can become complex for large-scale systems   |
| Performance  | REST is considered to have better performance due to its lightweight message formats and stateless nature.| SOAP messages tend to be larger due to XML formatting, which can lead to higher network overhead and potentially slower performance |
| Flexibility | REST provides more flexibility in terms of data formats and doesn't impose strict rules on how data should be structured.| SOAP has a strict structure defined by its XML schema, which can limit flexibility. |
| Security  | REST relies on underlying protocols for security (HTTPS) and can be secured through various authentication methods.    | SOAP has built-in security features like WS-Security for message-level security.  |
| Implementation Complexity  | REST APIs are generally simpler to implement due to their use of standard HTTP methods and lightweight messages.|  SOAP implementations can be more complex due to the strict XML structure and additional features.  |
