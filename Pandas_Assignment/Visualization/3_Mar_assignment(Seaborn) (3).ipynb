### Q1 : Name any five plots that we can plot using the Seaborn library.Also, state <br>the uses of each plot.?
### Ans :-
<b>Visualization Using Seaborn :</b><br>
- <b>Visualization Is A Representation Of Data Or Information By Using Charts,<b><br>Graphs And Maps. It Provides An Easy Way To Understand Patterns, Outliers And Trends Hidden In The Data Which Cannot Be Understood By Simply Looking At The Data Tables.</b><br><br>
- <b>Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface <br>for drawing attractive and informative statistical graphics.</b>

### (1) Seaborn Bar plot :<br>
- **Bar plots are used to visualize the relationship between a categorical variable and a continuous variable. In a bar plot, each bar represents the mean or median (or any aggregation) of the continuous variable for each category.<br> <br>**
- **In Seaborn, bar plots can be created using the barplot() function.**
### Example :

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
plt.figure(figsize=(3,2))
sns.barplot(x="class", y="fare", hue="sex", errorbar=None, palette="muted", data=titanic)

plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Average Fare by Class and Gender on the Titanic")
# display the plot
plt.show()
```

### (2) Seaborn histograms :
- **Histograms visualize the distribution of a continuous variable. In a histogram, the data is divided into bins and the height of each bin represents the frequency or count of data points within that bin.<br>**
-  **In Seaborn, histograms can be created using the histplot() function.**
### Example :

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
plt.figure(figsize=(3,2))
sns.histplot(x="petal_length", data=iris)
```

### (3) Seaborn density plots :
- **Density plots, also known as kernel density plots, are a type of data visualization that display the distribution of a continuous variable. They are similar to histograms, but instead of representing the data as bars, density plots use a smooth curve to estimate the density of the data.<br>**
- **In Seaborn, density plots can be created using the kdeplot() function.** 

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

plt.figure(figsize=(3,2))
sns.kdeplot(data=tips, x="total_bill")
```

### (4) Seaborn violin plots :
- **A violin plot is a type of data visualization that combines aspects of both box plots and density plots. It displays a density estimate of the data, usually smoothed by a kernel density estimator, along with the interquartile range (IQR) and median in a box plot-like form.<br>** 

- **The width of the violin represents the density estimate, with wider parts indicating higher density, and the IQR and median are shown as a white dot and line within the violin.**
### Example :

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

plt.figure(figsize=(4,3))
sns.violinplot(x="species", y="petal_length", data=iris)

plt.show()
```

### (5) Seaborn heatmaps :
- **A heatmap is a graphical representation of data that uses colors to depict the value of a variable in a two-dimensional space. Heatmaps are commonly used to visualize the correlation between different variables in a dataset.**
### Example :

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')

flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum', observed=False)

plt.figure(figsize=(6,4))
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d')

plt.title('Passengers per month')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```

### Q 2 : Load the "fmri" dataset using the load_dataset function of seaborn.<br> Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.<br><br>Note: timepoint, signal, event, and region are columns in the fmri dataset.
### Ans :-
**Seaborn line plots<br>**
**Line plots are used to visualize trends in data over time or other continuous variables. In a line plot, each data point is connected by a line, creating a smooth curve.<br>**
**In Seaborn, line plots can be created using the lineplot() function.<br><br>**
### fmri Dataset in Seaborn<br>
- **The "fmri" dataset in Seaborn is a sample dataset that contains functional magnetic resonance imaging (fMRI) data. fMRI is a neuroimaging technique that measures brain activity by detecting changes associated with blood flow. The "fmri" dataset in Seaborn typically includes time series data representing brain activity over time**

```python
import seaborn as sns

import matplotlib.pyplot as plt




fmri = sns.load_dataset("fmri")

plt.figure(figsize=(6,4))

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False, data=fmri)

plt.xlabel("Timepoint")

plt.ylabel("Signal Intensity")

plt.title("Changes in Signal Intensity over Time")

plt.show()
```

### Q 3 : Load the "titanic" dataset using the load_dataset function of seaborn.<br><br> Plot two box plots using x ='pclass', y='age' and y = 'fare'.<br><br>Note: pclass, age, and fare are columns in the titanic dataset.
### Ans :- 
**Seaborn box plots<br>**
- **Box plots are a type of visualization that shows the distribution of a dataset. They are commonly used to compare the distribution of one or more variables across different categories.<br><br>**
- **It gives the information about the maximum, minimum, mean, first quartile, and third quartile of a continuous variable. Also, it equips us with knowledge of outliers.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(8, 5))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot of Fare by Pclass')

plt.show()
```

### Q 4 : Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
### Ans :-

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

plt.figure(figsize=(6, 5))
sns.histplot(data=diamonds, x='price', hue='cut', stat='density', common_norm=False, kde=False)

plt.xlabel('Price')
plt.ylabel('Density')
plt.title('Histogram of Diamond Prices by Cut')

plt.show()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
### Ans :-
**Seaborn pair plots<br>**
- **Pair plots are a type of visualization in which multiple pairwise scatter plots are displayed in a matrix format. Each scatter plot shows the relationship between two variables, while the diagonal plots show the distribution of the individual variables.**

```python
import seaborn as sns

import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.pairplot(data=iris, hue="species", diag_kind="kde", palette="husl")

plt.title("Iris Dataset Pair Plot")

plt.show()
```
